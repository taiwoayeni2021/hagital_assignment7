```python
import pandas as pd
```

```python
netflix = pd.read_csv(r"C:\Users\user\Desktop\HAGITAL\5. DATA ENGINEERING\ASSIGNMENTS\netflix_titles.csv")
```

```python
netflix.head()
```

```python
# Fill missing values in 'country' column with 'Unknown'
netflix['country'] = netflix['country'].fillna('Unknown')
```

```python
# Verify that there are no more missing values in 'country'
netflix['country'].isna().sum()
```

```python
# Drop rows where 'title' or 'type' is missing
netflix = netflix.dropna(subset=['title', 'type'])
```

```python
# Check for missing values in 'title' and 'type'
netflix[['title', 'type']].isna().sum()
```

```python
# Count Movies vs TV Shows
netflix['type'].value_counts()
```

```python
# Top 10 most frequent countries
netflix['country'].value_counts().head(10)
```

```python
# Convert 'date_added' column to datetime format
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
```

```python
netflix.head()
```

```python
# Create new column 'year_added' from 'date_added'
netflix['year_added'] = netflix['date_added'].dt.year
```

```python
netflix[['date_added', 'year_added']].head()
```

```python
# Filter shows released before the year 2000
shows_before_2000 = netflix[netflix['release_year'] < 2000]
```

```python
shows_before_2000.head()
```

```python
# count films released before the year 2000
len(shows_before_2000)
```

```python
# Split the genres and count all occurrences
from collections import Counter
```

```python
# Split and flatten the genre list
all_genres = netflix['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]
```

```python
# Count the most common genre
genre_counts = Counter(flat_genres)
most_common_genre = genre_counts.most_common(1)

most_common_genre
```

```python
# Filter shows directed by 'Rajkumar Hirani'
hirani_shows = netflix[netflix['director'] == 'Rajkumar Hirani']
```

```python
# Display the result
hirani_shows
```

```python
# Count shows by rating
netflix['rating'].value_counts()
```

```python
# Filter only movies
movies = netflix[netflix['type'] == 'Movie'].copy()
```

```python
# Extract duration as integer (remove 'min' and convert to numeric)
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)
```

```python
# Filter movies with duration > 100 minutes
long_movies = movies[movies['duration_minutes'] > 100]
```

```python
# Show result
long_movies[['title', 'duration', 'duration_minutes']]
```

```python
# Extract numeric duration and create 'content_length' column
netflix['content_length'] = netflix['duration'].str.extract('(\d+)').astype('Int64')
```

```python
# Preview the result
netflix[['duration', 'content_length']].head()
```

```python
# Sort by release_year in descending order
netflix_sorted = netflix.sort_values(by='release_year', ascending=False)
```

```python
# To view a few rows
netflix_sorted[['title', 'release_year']].head()
```
